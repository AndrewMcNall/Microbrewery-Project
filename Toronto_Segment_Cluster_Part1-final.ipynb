# Segmenting and Clustering Neighborhoods in Toronto - Part 1

Import libraries needed for project.

```python
import pandas as pd
from pandas.io.html import read_html
```

Save Toronto zip code website to variable 'url'. Read the table data and save to variable 'table'.

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

table = read_html(url, attrs = {'class' : 'wikitable'})
```

Drop rows in which zip codes have no Boroughs assigned. Print dataframe of Postal Code, Boroughs and Neighborhoods.

```python
Tcodes = table[0].dropna(axis = 0, how = 'any').reset_index(drop = True)

Tcodes.style
```
